```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_log(filename):
    with open(filename) as f:
        lines = f.readlines()
    lines = [l for l in lines if l.startswith('SUMMARY')]
    lines = [l.split(' ') for l in lines]
    lines = [[w.strip(';') for w in l] for l in lines]
    
    iters = []
    res_dqn = []
    # for i,s in enumerate(lines[0]):
    #     print(i,s)
    
    for l in lines:
        iters.append(int(l[4]))
        res_dqn.append(float(l[12]))
        # print(f'{int(l[4]): 6d} {float(l[12]):6.2f}')
    
    return iters, lines, res_dqn
        
    # plt.ylim(top=4, bottom=2)

    # DQN结果
    plt.plot(iters, res_dqn, label='DQN-BS-BS-BS')

    # RND结果
    #plt.axhline(y=float(lines[0][20]), color='g', linestyle='--', label='RND-BS-BS-BS')

    # RND结果
    #plt.axhline(y=float(lines[0][28]), color='b', linestyle='--', label='STRM-BS-BS-BS')

    # BS结果
    plt.axhline(y=float(lines[0][-1]), color='r', linestyle='--', label='BS-BS-BS-BS')

    plt.legend()

iter11, lines11, res11 = plot_log('exp1log1')
iter12, lines12, res12 = plot_log('exp1log2')
iter13, lines13, res13 = plot_log('exp1log3')

# DQN结果
plt.plot(iter11, res11, label='DQN-BS-BS-BS')
# BS结果
plt.axhline(y=float(lines11[0][-1]), color='g', linestyle='--', label='BS-BS-BS-BS')

# DQN结果
plt.plot(iter12, res12, label='DQN-STRM-STRM-STRM')
# BS结果
plt.axhline(y=float(lines12[0][-1]), color='r', linestyle='--', label='BS-STRM-STRM-STRM')

# DQN结果
plt.plot(iter13, res13, label='DQN-RND-RND-RND')
# BS结果
plt.axhline(y=float(lines13[0][-1]), color='b', linestyle='--', label='BS-RND-RND-RND')

plt.legend()
plt.show()
```

```python
iter21, lines21, res21 = plot_log('exp1log1')
iter22, lines22, res22 = plot_log('exp2log2')
iter23, lines23, res23 = plot_log('exp2log3')
iter24, lines24, res24 = plot_log('exp2log4')

# plt.ylim(bottom=2, top=4)

# BS结果
plt.axhline(y=float(lines21[0][-1]), color='r', linestyle='--', label='BS-BS-BS-BS')

# DQN结果
plt.plot(iter21, res21, label='DQN-BS-BS-BS')

# DQN结果
plt.plot(iter22, res22, label='BS-DQN-BS-BS')

# DQN结果
plt.plot(iter23, res23, label='BS-BS-DQN-BS')

# DQN结果
plt.plot(iter24, res24, label='BS-BS-BS-DQN')

plt.legend()
plt.show()
```
